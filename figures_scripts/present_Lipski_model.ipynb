```python
def get_super(x):
    normal = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz0123456789+-=()"
    super_s = "ᴬᴮᶜᴰᴱᶠᴳᴴᴵᴶᴷᴸᴹᴺᴼᴾQᴿˢᵀᵁⱽᵂˣʸᶻᵃᵇᶜᵈᵉᶠᵍʰᶦʲᵏˡᵐⁿᵒᵖ۹ʳˢᵗᵘᵛʷˣʸᶻ⁰¹²³⁴⁵⁶⁷⁸⁹⁺⁻⁼⁽⁾"
    res = x.maketrans(''.join(normal), ''.join(super_s))
    return x.translate(res)
```

```python
Lipski_model_cue_cpeech = loadmat("processed_data/Lipski_model_cue_cpeech.mat")

encoder_r2_cue = Lipski_model_cue_cpeech['encoder_r2_cue']
encoder_num_elec_cue = Lipski_model_cue_cpeech['encoder_num_elec_cue']
nonencoder_r2_cue = Lipski_model_cue_cpeech['nonencoder_r2_cue']
nonencofer_num_elec_cue = Lipski_model_cue_cpeech['nonencofer_num_elec_cue']

encoder_r2_speech = Lipski_model_cue_cpeech['encoder_r2_speech']
encoder_num_elec_speech = Lipski_model_cue_cpeech['encoder_num_elec_speech']
nonencoder_r2_speech = Lipski_model_cue_cpeech['nonencoder_r2_speech']
encoder_r2_cue = Lipski_model_cue_cpeech['encoder_r2_cue']
```

```python
fig, axs = plt.subplots(1, 1, constrained_layout=True, figsize=(5, 5), dpi=200)
axs.patch.set_alpha(0)

num_bins = 80
axs.hist(list(nonencoder_r2_cue), bins=num_bins, label='Non encoder')
axs.hist(list(encoder_r2_cue), bins=num_bins, label='Encoder')
axs.set_xlim(-0.05, 1.05)
axs.legend(fontsize=20, frameon=0)
axs.set_xticklabels([round(item, 2) for item in axs.get_xticks()], fontsize=25)
axs.set_yticklabels(axs.get_yticks().astype(int), fontsize=25)
axs.set_ylabel('Number of units', fontsize=25)
axs.set_xlabel(f"R{get_super('2')}", fontsize=25)

plt.show();
```
